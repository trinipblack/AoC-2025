```python
raw_data_test = [
"L68",
"L30",
"R48",
"L5",
"R60",
"L55",
"L1",
"L99",
"R14",
"L82"
]
```

```python
dial_initial_start = 50
total_dial_positions = 100

def read_rotation(instruction):
    direction = instruction[0]
    steps = int(instruction[1:])
    if direction == 'R':
        return steps
    elif direction == 'L':
        return -steps
    else:
        raise ValueError("Invalid instruction format")

def calculate_single_rotation(rotation, total_dial_positions, dial_start):
    raw_sum = (dial_start + rotation)
    times_passing_through_zero = 0
    final_position = raw_sum % total_dial_positions
    if raw_sum <= 0 or raw_sum >= total_dial_positions:
        if raw_sum == 0 or raw_sum == 100:
            times_passing_through_zero = 1
        else:
            times_passing_through_zero = abs(raw_sum) // total_dial_positions
        if raw_sum < 0 and dial_start != 0:
            times_passing_through_zero += 1
        # print(f'Passing through zero position because raw sum {dial_start} + {rotation} = {raw_sum}, times: {times_passing_through_zero}')
    return final_position, times_passing_through_zero

def calculate_final_position(instructions, total_dial_positions, dial_start):
    current_position = dial_start
    # print(f"Starting dial position: {current_position}")
    dial_positions_log = [current_position]
    total_times_passing_through_zero = 0
    for instruction in instructions:
        rotation = read_rotation(instruction)
        current_position, times_passing_through_zero = calculate_single_rotation(rotation, total_dial_positions, current_position)
        total_times_passing_through_zero += times_passing_through_zero
        # print(f"After instruction {instruction}, dial position is {current_position}")
        dial_positions_log.append(current_position)
    return dial_positions_log, total_times_passing_through_zero

def find_number_of_positions(dial_number, dial_positions_log):
    return dial_positions_log.count(dial_number)
```

```python
# Test
data_to_use = raw_data_test
dial_positions_log, total_times_passing_through_zero = calculate_final_position(data_to_use, total_dial_positions, dial_initial_start)
print(f'Dial ending on 0: {find_number_of_positions(0, dial_positions_log)} times')
print(f'Total times passing through zero: {total_times_passing_through_zero}')
```
