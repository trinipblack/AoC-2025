```python
from time import time
```

```python
raw_data = open('files/input_day4.txt').read()
```

```python
example = '''..@@.@@@@.
@@@.@.@.@@
@@@@@.@.@@
@.@@@@..@.
@@.@@@@.@@
.@@@@@@@.@
.@.@.@.@@@
@.@@@.@@@@
.@@@@@@@@.
@.@.@@@.@.'''
```

```python
def process_data(data):
    return data.strip().split('\n')
rolls_map_example = process_data(example)
rolls_map = process_data(raw_data)
rolls_map_example
```

```python
def get_adjacent_single_position(rolls_map, x, y):
    x_len = len(rolls_map[0])
    y_len = len(rolls_map)
    ul = (x-1, y-1)
    u = (x, y-1)
    ur = (x+1, y-1)
    l = (x-1, y)
    r = (x+1, y)
    bl = (x-1, y+1)
    b = (x, y+1)
    br = (x+1, y+1)
    total_rolls = 0
    adjacent = [ul, u, ur, l, r, bl, b, br]
    for ax, ay in adjacent:
        if ax >= x_len or ax < 0 or ay >= y_len or ay < 0:
            continue
        else:
            if rolls_map[ay][ax] == '@':
                total_rolls += 1
    return total_rolls
```

```python
get_adjacent_single_position(rolls_map_example, 0, 0)
```

```python
def get_accessible_rolls(rolls_map, n_max_adj_rolls = 4):
    total_accessible = 0
    for y, row in enumerate(rolls_map):
        for x, position in enumerate(row):
            is_roll = position == '@'
            if not is_roll:
                continue
            total_adjacent = get_adjacent_single_position(rolls_map, x, y)
            if total_adjacent < n_max_adj_rolls:
                total_accessible += 1
    return total_accessible
accessible_rolls = get_accessible_rolls(rolls_map_example)
print(f'Accessible rolls: {accessible_rolls}')
```

```python
start = time()
accessible_rolls = get_accessible_rolls(rolls_map)
print(f'Accessible rolls: {accessible_rolls}')
end = time()
print(f'Part 1 took: {(end-start)*1000:.2f}ms')
```

## Part 2

```python
def remove_accessible_rolls(rolls_map, n_max_adj_rolls = 4):
    new_rolls_map = rolls_map.copy()
    total_accessible = 0
    for y, row in enumerate(rolls_map):
        for x, position in enumerate(row):
            is_roll = position == '@'
            if not is_roll:
                continue
            total_adjacent = get_adjacent_single_position(rolls_map, x, y)
            if total_adjacent < n_max_adj_rolls:
                total_accessible += 1
                new_rolls_map[y] = new_rolls_map[y][:x] + 'x' + new_rolls_map[y][x+1:]
    return total_accessible, new_rolls_map
```

```python
total_accessible, new_rolls_map = remove_accessible_rolls(rolls_map_example)
```

```python
def recursive_removal_of_accessible_rolls(rolls_map, n_max_adj_rolls = 4):
    total_accessible, new_rolls_map = remove_accessible_rolls(rolls_map)
    total_removed = total_accessible
    while total_accessible > 0:
        total_accessible, new_rolls_map = remove_accessible_rolls(new_rolls_map)
        total_removed += total_accessible
    return total_removed, new_rolls_map
    
total_removed, new_rolls_map = recursive_removal_of_accessible_rolls(rolls_map_example)
print(f'Total removed: {total_removed}')
new_rolls_map
```

```python
start = time()
total_removed, new_rolls_map = recursive_removal_of_accessible_rolls(rolls_map)
print(f'Total removed: {total_removed}')
end = time()
print(f'Part 2 took: {(end-start)*1000:.2f}ms')
```
