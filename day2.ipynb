```python
example = '11-22,95-115,998-1012,1188511880-1188511890,222220-222224,1698522-1698528,446443-446449,38593856-38593862,565653-565659,824824821-824824827,2121212118-2121212124'
example
```

```python
def process_data(data):
    ranges = data.strip().split(',')
    processed = [(int(r.split('-')[0]), int(r.split('-')[1])) for r in ranges]
    return processed

def get_ranges(processed_data):
    ranges = []
    for start, end in processed_data:
        ranges.append(list(range(start, end + 1)))
    return ranges
    
get_ranges(process_data(example))
```

```python
def search_invalid_in_single_range(range_list):
    invalid_numbers = [] # numbers where first half == second half
    for number in range_list:
        str_num = str(number)
        length = len(str_num)
        if length % 2 == 0:
            mid = length // 2
            first_half = str_num[:mid]
            second_half = str_num[mid:]
            if first_half == second_half:
                invalid_numbers.append(number)
        else:
            continue

    return invalid_numbers
```

```python
def get_sum_of_invalid_numbers(data, problem_part_int=1):
    ranges = get_ranges(process_data(data))
    total_invalid_numbers = []
    for range_list in ranges:
        if problem_part_int == 2:
            invalids = search_invalid_in_single_range_part2(range_list)
        else:
            invalids = search_invalid_in_single_range(range_list)
        total_invalid_numbers.extend(invalids)
    return sum(total_invalid_numbers)
```
