```python
# data
raw_data = open('files/input_day3.txt').read()
# raw_data
```

```python
example = '''987654321111111
811111111111119
234234234234278
818181911112111'''
example
```

```python
def process_data(data):
    return data.strip().split('\n')
banks_example = process_data(example)
banks_problem = process_data(raw_data)
banks_example
```

```python
def find_largest_2d_num_single_bank(bank):
    maxn = max(bank[:-1])
    first_position_max = bank.index(maxn)
    max_of_rest = max(bank[first_position_max+1:])
    return 10*int(maxn) + int(max_of_rest)
find_largest_2d_num_single_bank(banks_example[3])
```

```python
def calculate_result(banks):
    max_joltage_list = [find_largest_2d_num_single_bank(bank) for bank in banks]
    return sum([int(j) for j in max_joltage_list])
calculate_result(banks_example)
```

```python
calculate_result(banks_problem)
```

## Part 2

```python
def find_largest_12d_num_single_bank(bank, total_len = 12):
    remaining_bank = bank
    n = ''
    n_len = len(n)
    while n_len < total_len:
        n_remaining = total_len - n_len
        # print(f'remaining bank: {remaining_bank} remaining n: {n_remaining}')
        maxn = max(remaining_bank[:-(n_remaining-1)]) if n_remaining > 1 else max(remaining_bank)
        # print(f'maxn: {maxn}')
        first_position_max = remaining_bank.index(maxn)
        # print(f'first_position_max: {first_position_max}')
        remaining_bank = remaining_bank[first_position_max+1:]
        n += maxn
        # print(f'number so far: {n}')
        n_len = len(n)
        if n_remaining + len(n) == total_len:
            n += remaining_bank
            break
    # print(f'final number: {n}, len: {len(n)}')
    return n
```

```python
def calculate_result_part2(banks):
    max_joltage_list = [find_largest_12d_num_single_bank(bank) for bank in banks]
    return sum([int(j) for j in max_joltage_list])
calculate_result_part2(banks_example)
```

```python
calculate_result_part2(banks_problem)
```
